# **Тема №1. Чтение, запись и отображение**



## **Задание 1:**  
Напишите процедуру, которая:

- Считывает цветное изображение в режиме `cv2.IMREAD_COLOR`.
- Преобразует его в оттенки серого с использованием `cv2.cvtColor()`.
- Сохраняет полученное черно-белое изображение на диск.
- Отображает оба изображения (оригинальное цветное и преобразованное черно-белое).




````python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
````

````python
def process_image(input_image_path, output_image_path):
  img = cv.imread(input_image_path, cv.IMREAD_COLOR)
  img_color = cv.cvtColor(img, cv.COLOR_BGR2RGB)
  img_gray = cv.cvtColor(img_color, cv.COLOR_RGB2GRAY)
  cv.imwrite(output_image_path, img_gray)

  fig, axes = plt.subplots(1, 2, figsize=(12, 6))

  axes[0].imshow(img_color)
  axes[0].set_title('Исходное изображение')

  axes[1].imshow(img_gray, cmap='gray')
  axes[1].set_title('Оттенки серого')

  for ax in axes:
    ax.axis('off')

  plt.tight_layout()
  plt.show()

input_image_path = '/content/klub-cloud9-iz-ssha-podpisal-rossijskij-sostav-gambit-po-cs-go-igrokov-vyvezut-iz-rossii_16508024591.png'  # Замените на путь к вашему изображению
output_image_path = '/content/gray_image.jpg'  # Замените на путь для сохранения ч/б изображения

process_image(input_image_path, output_image_path)
````
